```
import sys
import string
import re

copy_klas1=klas1
copy_klas2=klas2
Klas1=[]
Klas2=[]
frequency1 = {}
text_string=copy_klas1.lower()
match_pattern = re.findall(r'\b[а-я,ї,є,і,ґ,a-z,\’]{1,15}\b', text_string)

for word in match_pattern:
    count = frequency1.get(word,0)
    frequency1[word] = count + 1

frequency2 = {}
text_string=copy_klas2.lower()
match_pattern = re.findall(r'\b[а-я,ї,є,і,ґ,a-z,\’]{1,15}\b', text_string)

for word in match_pattern:
    count = frequency2.get(word,0)
    frequency2[word] = count + 1

frequency_list1 = frequency1.keys()
for words in frequency_list1:
    if words not in sw:
        Klas1.append(words)

frequency_list2 = frequency2.keys()
for words in frequency_list2:
    if words not in sw:
        Klas2.append(words)

Count_spam=len(Klas1)
Count_ham=len(Klas2)

for j in Klas1:
   if j in Klas2:
     Klas2.remove(j)

Klas=[]
Klas.extend(Klas1)
Klas.extend(Klas2)

#----------------------------------------
text_str=klas1.lower()
freq = {}
text_string=klas1.lower()
match_pattern = re.findall(r'\b[а-я,ї,є,і,ґ,a-z,\’]{1,15}\b', text_string)

for word in match_pattern:
    count = freq.get(word,0)
    freq[word] = count + 1
    
Freq_klas1=[]

for words in Klas:
    if words in Klas1:
        Freq_klas1.append(int(freq[words]))
    else:
        Freq_klas1.append(0)
        
data1 = {'Words':Klas,
        'Frequency':Freq_klas1}

df1 = pd.DataFrame(data1)
pd.options.display.max_rows = 1000
df1.head(1000)
```

```
text_str=klas2.lower()
freq = {}
text_string=klas2.lower()
match_pattern = re.findall(r'\b[а-я,ї,є,і,ґ,a-z,\’]{1,15}\b', text_string)

for word in match_pattern:
    count = freq.get(word,0)
    freq[word] = count + 1
    
Freq_klas2=[]

for words in Klas:
    if words in Klas2:
        Freq_klas2.append(int(freq[words]))
    else:
        Freq_klas2.append(0)
        
data2 = {'Words':Klas,
        'Frequency':Freq_klas2}

df2 = pd.DataFrame(data2)
pd.options.display.max_rows = 1000
df2.head(1000)
```

```
sentence=input('Введіть повідомлення: ')
Count_all=len(Klas)

def spam_ham(sen):
    freq={}
    text_string=sen.lower()
    list_words = re.findall(r'\b[а-я,ї,є,і,ґ,a-z,\’]{1,15}\b', text_string)
    for word in list_words:
        if word in sw:
            list_words.remove(word)

    list_prob_spam=[]
    list_prob_ham=[]

    for word in list_words:
        if word in Klas:
            spam=Freq_klas1[Klas.index(word)]
            ham=Freq_klas2[Klas.index(word)]
            list_prob_spam.append((1+spam)/(Count_all+Count_spam))
            list_prob_ham.append((1+ham)/(Count_all+Count_ham))
        else:
            list_prob_spam.append(1/(Count_all+Count_spam))
            list_prob_ham.append(1/(Count_all+Count_ham))

    sum_list_prob_spam=sum(list_prob_spam)
    sum_list_prob_ham=sum(list_prob_ham)
    print('Вірогідність потраплення до спаму: ', sum_list_prob_spam,'\nВірогідність потраплення до не спаму: ', sum_list_prob_ham)
    if sum_list_prob_ham > sum_list_prob_spam:
        print('Не спам')
    elif sum_list_prob_ham < sum_list_prob_spam:
        print('Спам')
    else:
        print('Неможливо визначити')

spam_ham(sentence)
```
